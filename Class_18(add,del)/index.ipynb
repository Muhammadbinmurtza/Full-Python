```python
def df_obtained(s1:int , s2 : int , s3 : int , s4 : int , s5 : int)-> tuple[int,int,float,str]:
    total : int = 500
    obtained_marks : int = s1 + s2 + s3 + s4 + s5
    per : float = obtained_marks / total * 100
    grade : str = ""
    if per >= 80:
        grade = "A+" 
    elif per >= 70:
        grade = "A"
    elif per >= 60:
        grade = "B"
    elif per >= 50:
        grade = "C"
    elif per >= 40:
        grade = "D"
    else:
        grade = "F"
    return total,obtained_marks,per,grade
df[["Total","Marks","Percentage","Grade"]] = df[['s1', 's2', 's3', 's4', 's5']].apply(lambda x : df_obtained(*x) , axis=1,
                                                                                                        result_type="expand")
df
```

```python
def all_columns(a1 : int , a2 : int , a3 : int , a4 : int , a5 : int)-> tuple[int,int,float,str]:
    TotalMarks : int = 500
    ObtainedMarks : int = a1+a2+a3+a4+a5
    Percentage : float = ObtainedMarks / TotalMarks * 100
    grade : str = ""
    if Percentage >= 80:
        grade = "A+" 
    elif Percentage >= 70:
        grade = "A"
    elif Percentage >= 60:
        grade = "B"
    elif Percentage >= 50:
        grade = "C"
    elif Percentage >= 40:
        grade = "D"
    elif Percentage >= 30:
        grade = "E"
    else:
        grade = "F"
    return TotalMarks,ObtainedMarks,Percentage,grade
column_mix[["TotalMarks","ObtainedMarks","Percentage","grade"]] = column_mix[['a1','a2','a3','a4','a5']].apply(lambda a : all_columns(*a) , axis= 1,
                                                                                                               result_type="expand")
column_mix

```
