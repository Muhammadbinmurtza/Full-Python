# PASS BY VALUE

````python
a : int = 5
def abc(num1:int)-> None:# immutable data type
    print(f"the value before {num1} and the adress is {id(num1)}")
    num1 = 6 #copy of num1 at new address 
    print(f"the value after assigning {num1} and the address is {id(num1)}")
abc(a) #original with original address
print(f"a value is {a}")
# you'll see the addresses has been changed
````

````python
a : list[int] = [1,2,3,4,5]
def abc(num1:list[int])-> None:# mutable data type
    num1.append(500)
    print(f"the value after assigning {num1} and the id is {id(num1)}")
abc(a)
print(f"a value is {a} and the id is {id(a)}")# address will be same bcz list is mutable object and address will not change bcz it changes original value
# passed by refrence ( means by nick name) it changes value in both local and global means mutable objects or data types can do this
````

````python
a : list[int] = [1,2,3,4,5]
def abc(num1:list[int])-> None:# mutable data type
    num1 = [7] # its adddres will change bcz we are assigning new value and it changes the address of this assigning
    print(f"the value is {num1} and the id is {id(num1)}")
    num1.append(500) # it will show 7,500 bcz we assigned the new value
    print(f"the value after assigning {num1} and the id is {id(num1)}")
abc(a)
print(f"a value is {a} and the id is {id(a)}")# address will be same bcz list is mutable object and address will not change bcz it changes original value
# passed by refrence ( means by nick name) it changes value in both local and global means mutable objects or data types can do this
````

## Creating own error

````python
class student():
    def __init__(self,roll_no:int,name:str,age:int) -> None:
        if age < 18 or age > 60:
            raise studentAgeError("you are not eligible")
        self.roll_no = roll_no
        self.name = name
        self.age = age

            

class studentAgeError(Exception):
    pass
student1 = student(8,'muhammad',36)
student2 = student(8,'muhammad',11)

print(student1.name,student1.roll_no,student1.age)
print(student2.name,student2.roll_no,student2.age)
````
